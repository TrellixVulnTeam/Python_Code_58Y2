```python
"""
Ass 09: Create a DataFrame from 3 different lists names, scoreA and score B having the 
same lengths. Store the DataFrame in to a csv file.
Read the csv file into another DataFrame then:
•  Find people having the highest score A.
•  Find people having the lowest score B.
•  Sort people in descending order of score A and score B respectively.
"""

np.random.seed(101)
cols_name = list("ABCDEF")
cols_scoreA = np.random.randn(6)
cols_scoreB = np.random.randn(6)

df = pd.DataFrame(data = list(zip(cols_name, cols_scoreA, cols_scoreB)), columns = ["Name", "Score_A", "Score_B"])
df.rename_axis("ID", inplace = True)
print("This is source dataframe.")
print(df)
df.to_csv("Ass9.csv")

print("*"*50)
print("Person having the highest score A.")
print(df.loc[df["Score_A"].idxmax()])

print("*"*50)
print("Person having the lowest score B.")
print(df.loc[df["Score_B"].idxmin()])

print("*"*50)
print("Sort people in descending order of score A and score B respectively.")
df.sort_values(by = ["Score_A", "Score_B"], inplace = True)
print(df)
```

```python
"""
Ass 12 : 
Read file euro_2012.csv into a DataFrame and check information by using info().
From the DataFrame:
•  Filter 5 columns Team, Goals, Yellow Cards, Red Cards and Players Used.  
•  Sort and display teams in the ascending order of 'Player Used'.
•  Sort and display teams in the ascending orders of 'Yellow Cards' and 'Red 
Cards' respectively
•  Calculate the mean of 'Yellow Cards' of all teams in Euro 2012
•  Filter teams having more than 9 goals and sort them in the descending order 
of 'Passes' 
•  Create a new column 'Shooting Accuracy' as ['Shots on target'] / (['Shots on 
target'] + ['Shots off target']) * 100
•  Sort and display teams, goals and shooting accuracy in the descending order 
of 'Shooting Accuracy'
•  Store the DataFrame into euro_12.csv with headers
"""

df = pd.read_csv("./Pandas Datafiles/euro_12.csv")
print("This is dataframe from euro_12")
df.rename_axis("ID", inplace = True)
print(df)
print("*"*50)

print("This is information of this dataframe.")
print(df.info())
print("*"*50)

print("Filer 5 columns : Team, Goals, Yellow Cards, Red Cards and Players Used.")
cols_1 = ["Team", "Goals", "Yellow Cards", "Red Cards", "Players Used"]
print(df[cols_1])
print("*"*50)

print("Sort and display teams in the ascending order of 'Player Used'.")
print(df.sort_values(by = "Players Used", ascending = True))
print("*"*50)

print("""Sort and display teams in the ascending orders of 'Yellow Cards' and 'Red 
Cards' respectively""")
print(df.sort_values(by = ["Yellow Cards", "Red Cards"]))

print("*"*50)
print("Calculate the mean of 'Yellow Cards' of all teams in Euro 2012")
print(df["Yellow Cards"].mean())

print("*"*50)
print("""Filter teams having more than 9 goals and sort them in the descending order 
of 'Passes' """)
df[df["Goals"] >= 9].sort_values(by = "Passes", ascending = False)

print("*"*50)
print("""Create a new column 'Shooting Accuracy' as ['Shots on target'] / (['Shots on 
target'] + ['Shots off target']) * 100""")
df["Shooting Accuracy"] = 100*df["Shots on target"]/(df["Shots on target"] + df["Shots off target"])
print(df)

print("*"*50)
print(""" Sort and display teams, goals and shooting accuracy in the descending order 
of 'Shooting Accuracy'""")
print(df[["Team", "Goals", "Shooting Accuracy"]].sort_values(by = "Shooting Accuracy", ascending = False))

df.to_csv("Ass12.csv", header = True)
```

```python
"""
Ass 14: 
Read file OfficeSupplies.csv into a DataFrame (A) with the column 'OrderDate' as 
the index. 
•  Check information by using info(), head() and tail().
•  Create the DataFrame (B): total sold units of each representative in each 
region.
•  Calculate maximum of total units in each region to store in a DataFrame (C)
•  Drop index of C.
•  Merge B and C into a DataFrame (D). 
•  Transform the DataFrame A and store it into a csv file similar to 
sales_min_max.csv
"""
df_A = pd.read_csv("./Pandas Datafiles/OfficeSupplies.csv")
print(df_A)
print("This is information of dataframe")
print(df_A.info())
print("*"*100)

print(df_A.head())
print("*"*100)

print(df_A.tail())
print("*"*100)

print("Create the DataFrame (B): total sold units of each representative in each region.")
df_B = df_A.groupby(["Region", "Rep"])["Units"].apply(lambda x: sum(x)).reset_index()
print(df_B)
print("*"*100)

print("Calculate maximum of total units in each region to store in a DataFrame (C)")

def f(x):
    max = x['Units'].max()
    x = x[x['Units'] == max]
    return x

df_C = df_A.groupby(['Region','Rep']).sum().reset_index()
df_C = df_C.groupby(['Region']).apply(lambda x: f(x))
print(df_C)
# print(_max)
# print(df_C)
print("*"*100)

print("Drop index of C")
"""I creat C have no index. hihi"""
df_C.reset_index(inplace =True, drop =True)
print(df_C)

"""
Merge B and C
"""
df_C = df_C.merge(df_B, how = "outer", left_index=True, right_index=True, on = "Region")
print(df_C)

"""
Transform This datafram and save files
"""
df_C.to_csv("Ass14.csv")
```

```python
"""
Ass 17: Write a program to generate data similar to the file sale_data.csv. Analyze data 
columns then use Numpy and Pandas to create random data.
"""

cols_name =np.random.choice(list("ABCDEFGHIK"), 2500)
cols_branches = np.random.choice(np.arange(2, 10), 2500)
cols_products = np.random.choice(np.arange(20, 200), 2500)

def random_dates(start, end, n, unit='D', seed=None):
    if not seed:  # from piR's answer
        np.random.seed(0)

    ndays = (end - start).days + 1
    return pd.to_timedelta(np.random.rand(n) * ndays, unit=unit) + start

start = pd.to_datetime('2015-01-01')
end = pd.to_datetime('2018-01-01')
time = list(random_dates(start, end, 2500))

df =pd.DataFrame(data = list(zip(cols_name, cols_branches, cols_products, time)), columns = ["Provides","Branches", "Product", "Dates"])
df
```
